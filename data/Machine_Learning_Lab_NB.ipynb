```python
df['Profit'] = [x.strip('$') for x in df['Profit']]
df['Profit'] = [x.replace(',', '') for x in df['Profit']]
df['Profit'] = [x.replace('$', '') for x in df['Profit']]
df['Profit'] = [float(x) for x in df['Profit']]
```

```python
df['Sales'] = [x.strip('$') for x in df['Sales']]
df['Sales'] = [x.replace(',', '') for x in df['Sales']]
df['Sales'] = [x.replace('$', '') for x in df['Sales']]
df['Sales'] = [float(x) for x in df['Sales']]
```

```python
df['month'] = pd.DatetimeIndex(df['Order.Date']).month
```

```python
def month_convert(b):
    list_ = []
    for x in df['month']:
        if x == 1:
            list_.append('January')
        elif x == 2:
            list_.append('February')
        elif x == 3:
            list_.append('March')
        elif x == 4:
            list_.append('April')
        elif x == 5:
            list_.append('May')
        elif x == 6:
            list_.append('June')
        elif x == 7:
            list_.append('July')
        elif x == 8:
            list_.append('August')
        elif x == 9:
            list_.append('September')
        elif x == 10:
            list_.append('October')
        elif x == 11:
            list_.append('November')
        elif x == 12:
            list_.append('December')
    return list_
```

```python
df['month'] = month_convert(df['month'])
```
